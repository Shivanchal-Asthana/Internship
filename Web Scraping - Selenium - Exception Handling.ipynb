```python
#creating empty list
comment=[]
comment_upvote=[]
time=[]

#fetching comments
cycle = 100
for i in range(cycle):
    try:
        com = driver.find_elements_by_xpath('//*[@id="content-text"]')
        for i in com:
            comment.append(i.text.replace('\n',''))
    except:
        comment.append('-')
        
#fetching comments upvoted
cycle = 100
for i in range(cycle):
    try:
        com_up = driver.find_elements_by_xpath('//*[@id="vote-count-middle"]')
        for i in com_up:
            comment_upvote.append(i.text.replace('\n',''))
    except:
        comment_upvote.append('-')
        
# fetching time
cycle = 100
for i in range(cycle):
    try:
        t = driver.find_elements_by_xpath('//*[@class="published-time-text above-comment style-scope ytd-comment-renderer"]')
        for i in t:
            time.append(i.text.replace('\n',''))
    except:
        time.append('-')
        
#Combining all the lists into a single dataframe
df_youtube=pd.DataFrame({'Comments': comment,
                'Comments_upvoted': comment_upvote,
               'Time': time,
                })
print("\t\t\tParam Sundari- Official Video")
df_youtube
```
